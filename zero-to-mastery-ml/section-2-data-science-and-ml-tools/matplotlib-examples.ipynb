```python
# 0. import matplotlib and get it readt for lotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x=[1,2,3,4]
y=[11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
        xlabel="x-axis",
        ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("../images/sample-plot.png")
```

## Making fig with NumPy arrays

* Line Plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplot

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data and create aline plot
fig, ax = plt.subplots()
ax.plot(x,x**2);
```

```python
# Use same data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x,x**2);
```

```python
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot form a dictionary
nut_butter_prices = {
    "Almond Butter": 10,
    "Peanut Butter": 8,
    "Cashew Butter": 12
}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Dan's Nut Butter Store",
        ylabel="Prices ($)");
```

```python
fig, ax = plt.subplots()
ax.barh( 
    list(nut_butter_prices.keys()), list(nut_butter_prices.values())
);
```
