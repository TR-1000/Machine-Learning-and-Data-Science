```python
def make_int(s):
    # remove '€'
    s = s.replace("€","")
    
    
    # if found, remove '.' and "M" or remove "M"or"K"
    if "." in s and "M" in s:
        s = s.replace(".","").replace("M","00000")
        
    else: 
        s = s.replace("M","000000").replace("K","000")
        
    
    #return value in interger form
    return int(s)
  
```

```python
value_df["Value"] = value_df["Value"].apply(make_int)
```

```python
value_df["Wage"] = value_df["Wage"].apply(make_int)
```
